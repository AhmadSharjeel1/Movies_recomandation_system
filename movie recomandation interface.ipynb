```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pywebio.input import *
from pywebio.output import *
from pywebio.platform.flask import webio_view
from flask import Flask
app = Flask(__name__)

column_names = ["movie id","movie title","release date","video release date","IMDb URL","unknown","Action","Adventure","Animation","Childrens","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western"]
item = pd.read_csv("u.item", sep = "|", names = column_names, encoding = "latin-1")

options_list = list(item["movie title"])

index_movie_dict = {}
for i in range(len(item)):
    index_movie_dict[i] = item.loc[i]['movie title']

movie_index_dict = {}
for i in range(len(item)):
    movie_index_dict[item.loc[i]['movie title']] = i

movie_release_date_dict = {}
for i in range(len(item)):
    movie_release_date_dict[item.loc[i]['movie title']] = item.loc[i]['release date']

item.drop(["movie id","movie title","release date","video release date","IMDb URL"], axis = 1, inplace = True)

from sklearn.metrics.pairwise import cosine_similarity
correlationMatrix = cosine_similarity(item)

item = pd.read_csv("u.item", sep='|', names = column_names, encoding='latin-1')
posters = pd.read_csv('movie_poster.csv', names = ['movie id', 'poster_link'])
df = posters.merge(item, on="movie id")
df = df.drop(columns=["unknown", "Action", "Adventure", "Animation", "Childrens", "Comedy", "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"])

movies_list1 = df['movie title']
poster_link = df['poster_link']
movie_poster_dict = {}
for i in range(len(df)):
    movie_poster_dict[movies_list1[i]] = poster_link[i]

def getStarted():
    img = open('Banner.jpeg', 'rb').read()
    put_image(img, width='100%')
    
    recently_watched_movie = select("Which movie have you watched recently? ", options = options_list, required = True)
    number_of_recommendations = input("How many recommendations do you need? ", type = NUMBER)
    put_markdown('Recommendations for you similar to `%s` movie are as follows:' % recently_watched_movie)
    
    # Generate personalized recommendations
    idx = movie_index_dict[recently_watched_movie]
    similarity_list = list(correlationMatrix[idx])
    lst = []
    for i in range(len(similarity_list)):
        lst.append((similarity_list[i],i))
    lst.sort(reverse = True)
    recommendation_list = []
    for element in lst:
        recommendation_list.append(index_movie_dict[element[1]])
    recommendation_list.remove(recently_watched_movie)
    final_recommendations = recommendation_list[:number_of_recommendations]
    for movie in final_recommendations:
        put_markdown('# `%s`' % movie)
        put_markdown('Release date = %s' % movie_release_date_dict[movie])
        # Display movie poster
        if(movie in movie_poster_dict.keys()):
            p_link = movie_poster_dict[movie]
            put_image(p_link, width = '1000px')
        else:
            put_text("Oops! Enough information not available.")
        put_html('<hr>')
    img = open('End Note.jpeg', 'rb').read()
    put_image(img, width='100%')
    
app.add_url_rule('/movierec', 'webio_view', webio_view(getStarted), 
                 methods=['GET', 'POST', 'OPTIONS'])
app.run(host='localhost', port=80)

# Check out http://localhost/movierec
```
